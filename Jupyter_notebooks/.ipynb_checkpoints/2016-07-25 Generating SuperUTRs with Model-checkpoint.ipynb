## Functions for performing simple evolution

For the sake of simplicity, we've only made point mutations -- i.e no sequence shuffling, deletions, insertions, etc.

Handy global variables:

```python
bases = ['A','C','G','T']
base_dict = dict(zip(bases, range(4)))
```

One hot encoding of sequences:

```python
def one_hot_encoding(seqs):
    n = len(seqs)
    total_width = 70
    X = np.zeros((n, 1, 4, total_width))
    
    for i in range(n):
        seq = seqs[i]
        
        for b in range(len(seq)):
            X[i, 0, base_dict[seq[b]], 10 + 50 - len(seq) + b] = 1.
            
    return X.astype(theano.config.floatX)
```

Make mutations in sequence:

```python
def make_mutation(seqs, pos, base):
    
    return [s[:pos] + base + s[pos + 1:] if pos < len(s) else s for s in seqs]
```

Evolve sequences:
    
Here we make all possible mutations to a given sequence, ask the model to score each, and take the best one. We do this for 40 rounds.

```python
def evolve_seqs(seqs, rounds = 5):
    all_seqs = {}
    all_scores = {}
    all_seqs[0] = seqs
    all_scores[0] = model.predict(one_hot_encoding(seqs)).flatten()
    
    for i in range(rounds):
        Y_mut = {}
        
        for b in bases:
            
            for p in range(50):
                mut_seqs = make_mutation(seqs, p, b)
                X_mut = one_hot_encoding(mut_seqs)
                Y_mut[b + '_' + str(p)] = model.predict(X_mut).flatten()
                best_muts = pd.DataFrame(Y_mut).apply(argmax, axis = 1).values
                
        seqs = [s[:int(m[2:])] + m[:1] + s[int(m[2:]) + 1:] for s, m in zip(seqs,best_muts)]
        all_seqs[i + 1] = seqs
        all_scores[i + 1] = model.predict(one_hot_encoding(seqs)).flatten()
        print i, pd.DataFrame(Y_mut).max(1).mean()
        
    return pd.DataFrame(all_seqs), pd.DataFrame(all_scores)

def make_random_seqs(n = 100):
    seqs = [''.join(np.random.choice(bases, 50)) for i in range(n)]
    
    return seqs
```

```python
native_cds = {}
with open('../ref/orf_coding.fasta') as f:
    line = f.readline()
    gene = line.split(' ')[0][1:]
    seq = ''
    while True:
        line = f.readline()
        if len(line)==0:
            break
        if line[0]=='>':
            native_cds[gene] = seq
            gene = line.split(' ')[0][1:]
            seq = ''
        else:
            seq += line[:-1]
native_cds = pd.Series(native_cds)
```

```python
yeast_tAI = pd.read_csv('../ref/tAI.tsv',sep='\t',index_col=0).Scer.to_dict()
yeast_tAI['TAA'] = 0.
yeast_tAI['TGA'] = 0.
yeast_tAI['TAG'] = 0.
```

```python
def score_gene_tai(seq):
    score = 0.
    for i in range(len(seq)/3-1):
        codon = seq[i*3:(i+1)*3]
        score += yeast_tAI[codon]
    return score/(len(seq)/3-1)

def score_gene_tai(seq):
    scores = []
    for i in range(len(seq)/3-1):
        codon = seq[i*3:(i+1)*3]
        scores.append(yeast_tAI[codon])
    return mean(scores)
```

```python
native_codon_scores = native_cds.apply(score_gene_tai)
```
